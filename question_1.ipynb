```python
# Write a python program to display all the header tags from wikipedia.org and make data frame.

from bs4 import BeautifulSoup
import requests
```

```python
wp = requests.get('https://en.wikipedia.org/wiki/Main_Page')
wp
```

```python
soup = BeautifulSoup(wp.content)
soup
```

```python
# All Header tag
header_tags = [] # empty list to store 
for i in soup.find_all('h2', class_="mp-h2"):
    header_tags.append(i.text.replace('\xa0...', ''))
    
header_tags
```

```python
# Making a DataFrame

import pandas as pd
df = pd.DataFrame({'HEARDER_TAGS':header_tags})
df
```
